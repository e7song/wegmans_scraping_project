```python
APP_ID  = "QGPPR19V8V"
API_KEY = "9a10b1401634e9a6e55161c3a60c200d"

# we’ll use the single‑index query endpoint for simplicity
URL = f"https://{APP_ID}-dsn.algolia.net/1/indexes/products/query"

headers = {
    "x-algolia-application-id": APP_ID,
    "x-algolia-api-key":        API_KEY,
    "Content-Type":             "application/json",
    "Origin":                   "https://www.wegmans.com",
    "Referer":                  "https://www.wegmans.com/",
    "User-Agent":               "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                                "AppleWebKit/537.36 (KHTML, like Gecko) "
                                "Chrome/138.0.0.0 Safari/537.36",
    "Accept":                   "application/json, text/plain, */*",
}

def make_params(store_id: int, page: int, hits_per_page: int = 1000):
    """
    Build the URL‑style params string that will match EVERY product
    in `store_id` that is sold in‑store.
    """
    return "&".join([
        "query=",                             # blank = match all products
        f"page={page}",                       # which page (0-based)
        f"hitsPerPage={hits_per_page}",       # max items per page
        "filters=" + "%20AND%20".join([
            f"storeNumber:{store_id}",        # only this store
            "isSoldAtStore:true",             # only items sold at store
            "fulfilmentType:instore"          # only instore fulfillment
        ])
    ])

# 1) Fetch page 0 to see how many pages there are
store = 68
body = {"params": make_params(store, page=0)}

resp = requests.post(URL, headers=headers, json=body)
resp.raise_for_status()

data0 = resp.json()
hits0     = data0["hits"]
total_hits = data0["nbHits"]
pages     = data0["nbPages"]

print(f"Store {store}: {total_hits} total hits over {pages} pages.")
all_hits = hits0

# 2) Loop the remaining pages
for p in range(1, pages):
    body["params"] = make_params(store, p)
    r = requests.post(URL, headers=headers, json=body)
    r.raise_for_status()
    page_hits = r.json()["hits"]
    print(f"  ▸ Page {p}: {len(page_hits)} hits")
    all_hits.extend(page_hits)

print(f"Collected {len(all_hits)} products for store {store}.")
```

```python
import requests

APP_ID    = "QGPPR19V8V"
API_KEY   = "9a10b1401634e9a6e55161c3a60c200d"
INDEX     = "products"
BASE_URL  = f"https://{APP_ID}-dsn.algolia.net/1/indexes/{INDEX}/query"

HEADERS = {
    "X-Algolia-API-Key":        API_KEY,
    "X-Algolia-Application-Id": APP_ID,
    "Content-Type":             "application/json"
}

# 1) Discover all facets and their counts
discover_payload = {
    "query": "",              # empty search
    "hitsPerPage": 0,         # we only want facets, no hits
    "facets": ["*"],          # ask for every facet
    "maxValuesPerFacet": 100  # up to 100 values per facet
}

resp = requests.post(BASE_URL, headers=HEADERS, json=discover_payload)
resp.raise_for_status()
data = resp.json()

facets = data.get("facets", {})
if not facets:
    print("⚠️  No facets returned. Make sure this index has attributesForFaceting set!")
else:
    print("Discovered facets:")
    for field, values in facets.items():
        print(f"\n▶ {field}:")
        for val, count in values.items():
            print(f"   • {val} ({count})")

# 2) Batch‐scrape through each facet value
all_hits = []
for field, values in facets.items():
    for val in values:
        print(f"\nFetching for {field} = {val!r} …")
        page = 0
        while True:
            batch_payload = {
                "query": "",
                "filters": f"{field}:\"{val}\"",
                "hitsPerPage": 1000,
                "page": page
            }
            r = requests.post(BASE_URL, headers=HEADERS, json=batch_payload)
            r.raise_for_status()
            page_data = r.json()
            hits = page_data.get("hits", [])
            all_hits.extend(hits)
            print(f"  • page {page}: {len(hits)} hits")
            if len(hits) < 1000:
                break
            page += 1

print(f"\n✅ Collected a total of {len(all_hits)} records.")
```

```python
import requests
from typing import List, Optional

API_KEY = "9a10b1401634e9a6e55161c3a60c200d"
APP_ID  = "QGPPR19V8V"
URL     = f"https://{APP_ID}-dsn.algolia.net/1/indexes/*/queries"

def get_brands_for_category(
    sub_id: str,
    store_id: int,
    session: Optional[requests.Session] = None
) -> List[str]:
    """
    Returns all the unique consumerBrandName facet values for the
    given store+subcategory, by asking Algolia for 0 hits and
    only facet counts.
    """
    sess = session or requests.Session()
    sess.headers.update({
        "X-Algolia-API-Key":        API_KEY,
        "X-Algolia-Application-Id": APP_ID,
        "Content-Type":             "application/json",
    })

    # Build the `params` string to:
    # 1) fetch 0 hits (we only want facet counts)
    # 2) filter by store & subcategory
    # 3) ask for consumerBrandName as a facet (double-encoded JSON array)
    params = (
        "query=&"
        "hitsPerPage=0&"
        f"filters=storeNumber:{store_id} AND category.key:{sub_id}&"
        "facets=%5B%22consumerBrandName%22%5D&"
        "maxValuesPerFacet=1000"
    )

    payload = {
        "requests": [
            { "indexName": "products", "params": params }
        ]
    }

    resp = sess.post(URL, json=payload)
    resp.raise_for_status()
    result = resp.json()["results"][0]

    # Now `result["facets"]` will exist, and include our brand counts:
    brands_dict = result.get("facets", {}).get("consumerBrandName", {})
    return list(brands_dict.keys())

# example usage:
session = requests.Session()
brands = get_brands_for_category("2952098", 156, session=session)
print("Found brands:", brands)
```
